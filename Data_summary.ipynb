```python
import pandas as pd

# Load the dataset and strip spaces from column names to avoid issues
data = pd.read_csv("IHME_GBD.csv")
data.columns = data.columns.str.strip()  # Remove any leading/trailing spaces from column names

# Check if 'location_name' exists and filter if it does
if 'location_name' in data.columns:
    filtered_data = data[data['location_name'] == 'United States of America']
    # Display the filtered dataset
    print(filtered_data)
    # Optionally save the filtered dataset to a new CSV file
    filtered_data.to_csv("filtered_IHME_GBD.csv", index=False)
else:
    print("'location_name' column not found in the dataset!")

# List of dataset file names
dataset_files = ["asthma_in_children.csv", "pm_data_2021.csv", "IHME_GBD.csv", "Respiratory_Death.csv"]

# Dictionary to store datasets and their descriptions
datasets = {}

for file in dataset_files:
    # Load dataset
    data = pd.read_csv(file)
    datasets[file] = data
    
    # Print dataset overview
    print(f"\nDataset: {file}")
    print("First few rows:")
    print(data.head())

    # Statistical summary
    print("\nStatistical Summary:")
    print(data.describe())
```
